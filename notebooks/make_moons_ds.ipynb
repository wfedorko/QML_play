```python
import numpy as np
import sklearn as sklearn
from sklearn.datasets import make_moons
import matplotlib.pyplot as plt
```

```python
data,labels=make_moons(n_samples=1000, noise=.15, random_state=0)
```

```python
plt.scatter(x=data[:,0],y=data[:,1],c=labels)
```

```python
np.savez_compressed("moons.npz",data=data,labels=labels)
```

```python
npzfile = np.load("moons.npz")
```

```python
data_l=npzfile['data']
```

```python
labels_l=npzfile['labels']
```

```python
plt.scatter(x=data_l[:,0],y=data_l[:,1],c=labels_l)
```
